**Step 04:**
- Perform the Statistical Outlier Detection (Z-score).
1. Z-Score = 0: A Z-score of 0 indicates that the data point is exactly at the mean of the distribution.
2. Positive Z-Score: A positive Z-score indicates that the data point is above the mean. For example, a Z-score of 1.5 means the data point is 1.5 standard deviations above the mean.
3. Negative Z-Score: A negative Z-score indicates that the data point is below the mean. For example, a Z-score of -2 means the data point is 2 standard deviations below the mean.
4. Magnitude of Z-Score: The magnitude of the Z-score shows how far away the data point is from the mean. A larger absolute value of the Z-score indicates that the data point is farther from the mean, while a smaller absolute value indicates it is closer.

**Common Thresholds**.
1. Z-Score > 2 or < -2: Often considered unusual or significant, indicating the data point is more than 2 standard deviations away from the mean.
2. Z-Score > 3 or < -3: Typically considered an outlier, suggesting the data point is extremely far from the mean.
